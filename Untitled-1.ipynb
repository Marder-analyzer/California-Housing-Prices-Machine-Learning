# 1Ô∏è‚É£ What is PySpark and Why Do We Use It?

PySpark is Apache Spark's Python API used to work with big data. Normally, libraries like pandas or scikit-learn run in RAM and slow down when dealing with big data. PySpark is fast because it can work with big data in a distributed manner.

In summary:
- ‚úÖ It can process large data sets.
- ‚úÖ It offers powerful functions for machine learning (Spark ML).
- ‚úÖ It can train large models quickly by doing distributed processing.

Next we will start the Spark environment:

```python
# This is a required step for PySpark to work. You will need to do this first in every PySpark project.
from pyspark.sql import SparkSession

# Start the spark session
spark = SparkSession.builder.appName("HousingPricePrediction").getOrCreate()
```

# 3Ô∏è‚É£ Reading Data with PySpark


```python
# Read dataset
df = spark.read.csv("housing.csv", header=True, inferSchema=True)

# Show first few lines
df.show(5)
```

### Here:

+ header=True ‚Üí Takes the first row as column names.

+ inferSchema=True ‚Üí Automatically detects the data types of the columns (determines whether they are numeric or categorical).

Checking for missing data:

```python
from pyspark.sql.functions import col, isnan, when, count
df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
# Here, we filled in the missing total_bedrooms values ‚Äã‚Äãwith mean.
df = df.fillna({'total_bedrooms': df.selectExpr("mean(total_bedrooms)").collect()[0][0]})
```

# 6Ô∏è‚É£ Converting Categorical Data to Numbers

There are categorical variables in the dataset, such as ocean_proximity. We will convert it to numeric with StringIndexer:

```python
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol="ocean_proximity", outputCol="ocean_proximity_index")
df = indexer.fit(df).transform(df)
df.show(5)
```

# 7Ô∏è‚É£ Preparing Features for the Model (Feature Engineering)

Machine learning models in PySpark ML accept vectors. So we need to combine all the arguments into a single column. We will do this with VectorAssembler:

```python
from pyspark.ml.feature import VectorAssembler

# Columns to be used
feature_columns = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'ocean_proximity_index']

# Combine properties into a single column
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df = assembler.transform(df)

df.select("features").show(5, truncate=False)
```

# 8Ô∏è‚É£ Training a Machine Learning Model with PySpark

```python
from pyspark.ml.regression import LinearRegression, DecisionTreeRegressor, RandomForestRegressor, GBTRegressor

# Split the data into train-test
train_data, test_data = df.randomSplit([0.8, 0.2], seed=42)

# Linear Regression model
lr = LinearRegression(featuresCol="features", labelCol="median_house_value")
lr_model = lr.fit(train_data)

# Decision Tree model
dt = DecisionTreeRegressor(featuresCol="features", labelCol="median_house_value")
dt_model = dt.fit(train_data)

# Random Forest model
rf = RandomForestRegressor(featuresCol="features", labelCol="median_house_value", numTrees=10)
rf_model = rf.fit(train_data)

# Gradient Boosting model
gbt = GBTRegressor(featuresCol="features", labelCol="median_house_value")
gbt_model = gbt.fit(train_data)
```

# 9Ô∏è‚É£ Evaluating the Performance of the Model

```python
# Linear Regression estimates
lr_predictions = lr_model.transform(test_data)
lr_predictions.select("median_house_value", "prediction").show(10)

# Decision Tree predictions
dt_predictions = dt_model.transform(test_data)
dt_predictions.select("median_house_value", "prediction").show(10)

# Random Forest predictions
rf_predictions = rf_model.transform(test_data)
rf_predictions.select("median_house_value", "prediction").show(10)

# Gradient Boosting predictions
gbt_predictions = gbt_model.transform(test_data)
gbt_predictions.select("median_house_value", "prediction").show(10)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

# Define evaluator
rmse_evaluator = RegressionEvaluator(labelCol="median_house_value", predictionCol="prediction", metricName="rmse")
r2_evaluator = RegressionEvaluator(labelCol="median_house_value", predictionCol="prediction", metricName="r2")

# Linear Regression evaluation
lr_rmse = rmse_evaluator.evaluate(lr_predictions)
lr_r2 = r2_evaluator.evaluate(lr_predictions)

# Decision Tree evaluation
dt_rmse = rmse_evaluator.evaluate(dt_predictions)
dt_r2 = r2_evaluator.evaluate(dt_predictions)

# Random Forest evaluation
rf_rmse = rmse_evaluator.evaluate(rf_predictions)
rf_r2 = r2_evaluator.evaluate(rf_predictions)

# Gradient Boosting evaluation
gbt_rmse = rmse_evaluator.evaluate(gbt_predictions)
gbt_r2 = r2_evaluator.evaluate(gbt_predictions)

# Print the results to the screen
print(f"Linear Regression -> RMSE: {lr_rmse:.4f}, R¬≤: {lr_r2:.4f}")
print(f"Decision Tree -> RMSE: {dt_rmse:.4f}, R¬≤: {dt_r2:.4f}")
print(f"Random Forest -> RMSE: {rf_rmse:.4f}, R¬≤: {rf_r2:.4f}")
print(f"Gradient Boosting -> RMSE: {gbt_rmse:.4f}, R¬≤: {gbt_r2:.4f}")
```
